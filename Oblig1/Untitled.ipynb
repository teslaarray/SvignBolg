```python
import numpy as np
import matplotlib.pyplot as pl

k = 8
m = 2
A = 1.077032961
phi = 1.1902899

omega = 2  # sqrt(4)


def bev(t):
    return A * np.cos(omega * t + phi)


def speedo(t):
    return -A * omega * np.sin(omega * t + phi)


T = 2 * np.pi / omega

t = np.linspace(0, T, 50)
x = bev(t)
v = speedo(t)
p = v * m

pl.plot(x, p)
pl.axis("equal")
pl.xlabel("position (m)")
pl.ylabel("momentum (kg m/s)")
pl.show()
# får en ellipse

# ------------skalering----------------

# skalerer x aksen med maks amplitude av x(t)
# y aksen med ( maks amplitude av v(t) )*massen

xstj = x / A
pstj = p / (A * omega * m)

pl.plot(xstj, pstj)
pl.axis("equal")
pl.xlabel("position")
pl.ylabel("momentum")
pl.show()

# får en sirkel
```

```python
# en funksjon på formen abs ( cos(t) ) ser riktig ut

# dropper ballen fra 1m høyde

# tiden det tar å nå bakken
# 1/2 at**2 = 1m
t1 = np.sqrt(2 * 1 / 9.81)

# etter tiden t1 vil vi ha pi/2 i cos funksjonen, 2pi blir da 4*t1, som blir perioden
omega = 2 * np.pi / (4 * t1)

t = np.linspace(0, 5, 1000)
x = np.abs(np.cos(omega * t))

pl.plot(t, x)
pl.xlabel("tid (s)")
pl.ylabel("høyde (m)")
pl.show()
```

```python
# v(t) = -omega*sin(omega*t)

m = 1 #kg

# også glemmer vi retningen på mv og ser bare på abs (mv)
p = np.abs(-omega*np.sin(omega*t)*m)
pl.plot(t, p)
pl.show()
```

Og dette kan ikke være riktig, v(0.45) skal være rundt 4.4m/s, ikke 3.5, prøver noe annet

```python
g = -9.81
v = np.zeros_like(t)
s = np.zeros_like(t)

#v[0] = 0
s[0] = 1

dt = t[1]-t[0]
for i in range(len(t)-1):
    v[i+1] = v[i] + dt*g
    s[i+1] = s[i] + dt*v[i+1]
    v[i+1] *= np.sign(s[i+1]) # v endrer retning når ballen når bakken

pl.plot(t, v)
pl.ylabel("momentum ( kg m/s)")
pl.xlabel("tid (s)")
pl.show()
```

Nå ser det mer riktig ut, v(t) skal være lineær. Og vi får riktig fortegn.
