2b)

```python
import numpy as np
import matplotlib.pyplot as pl

memLen = 30e-3
celleLen = memLen / 3000


def br(l):
    return 0.1e-3 + (0.2e-3 / memLen) * l


def hy(l):
    return 0.3e-3 - (0.2e-3 / memLen) * l


def kf(l):
    return 1e-6 + ((0.1 - 1e-6) / memLen) * l


def rho(l):
    return 1500 + (1000 / memLen) * l


def m(l):
    return br(l) * hy(l) * celleLen * rho(l) # masse = volum * tetthet

def oscif(l):
    return np.sqrt(kf(l) / m(l))
```

2c)

```python
l = np.linspace(0, memLen, 3000)
omega0 = oscif(l)
pl.plot(l, omega0/(np.pi*2))
pl.xlabel("pos (m)")
pl.ylabel("frekvens (Hz)")
pl.show()
```

2d)

```python

M = m(l) # array av alle masser
c4 = 261.63
c4sh = 277.18


def getA(omf, b=0): # bruker formel fra boka
    return ((np.sqrt((omega0 ** 2 - omf ** 2) ** 2 + (b * omf / M) ** 2)) * M) ** -1


B = 1e-8 # prøvde flere verdier, denne ser bra ut

Ac4 = getA(2 * np.pi * c4, b=B) # amplituderespons for c4
Ac4sh = getA(2 * np.pi * c4sh, b=B) # for c4#

N = 100
pl.plot(l[0:N], Ac4[0:N], label="c4 amp resp")
pl.plot(l[0:N], Ac4sh[0:N], label="c4sh amp resp")
pl.legend()
pl.show()
```

2e)

```python
# her er det 2 Qer, en for hver frekvens

# finner indexer

i1 = Ac4.argmax()
i2 = Ac4sh.argmax()

# bruker formel for Q

K = kf(l)

Q1 = np.sqrt((M[i1]*K[i1])/B**2)
Q2 = np.sqrt((M[i2]*K[i2])/B**2)

print(f"Q1 = {Q1}\nQ2 = {Q2}")
```
